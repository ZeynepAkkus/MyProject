```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
import warnings
warnings.filterwarnings('ignore') 
```

#### 1. Draw the graphs below for the given data:

- Data:
- labels = ['UK', 'USA', 'GER', 'JAP', 'FRA']
- age_men = [40, 35, 45, 53, 28]
- age_women = [35, 32, 48, 48, 31]

```python
labels = ['UK', 'USA', 'GER', 'JAP', 'FRA']
age_men = [40, 35, 45, 53, 28]
age_women = [35, 32, 48, 48, 31]
```

### 1-Graph

```python
# functional
plt.bar(labels,age_men, width=0.35, label= "Men")
plt.title("Age means By Gender")

for i in range(len(age_men)):
    plt.text(x=i, y = age_men[i]+1,s = age_men[i],ha = "center")

plt.ylabel("Ages")

plt.legend();
plt.show()
```

```python
# OOP
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x , age_men, width, label='Men')

ax.set_ylabel('Ages')
ax.set_title('Age means By Gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(rects1, padding=3)

fig.tight_layout()

plt.show()
```

### 2-Graph

```python
# functional
plt.bar(labels,age_women, width=0.35, label= "Women")
plt.title("Age means By Gender")

for i in range(len(age_women)):
    plt.text(x=i, y = age_women[i]+1,s = age_men[i],ha = "center")

plt.ylabel("Ages")

plt.legend();
plt.show()
```

```python
# OOP
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects2 = ax.bar(x, age_women, width, label='Women')

ax.set_ylabel('Ages')
ax.set_title('Age means By Gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

### 3-Graph

```python
# OOP
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10,7))
rects1 = ax.bar(x - width/2, age_men, width, label='Men')
rects2 = ax.bar(x + width/2, age_women, width, label='Women')

ax.set_ylabel('Ages')
ax.set_title('Age means By Gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(rects1, padding=1)
ax.bar_label(rects2, padding=1)

fig.tight_layout()

plt.show()
```
