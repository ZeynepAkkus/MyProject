```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
print(sns.get_dataset_names())
```

```python
mpg = sns.load_dataset("mpg")
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
#describe all features, both numerical and categorical

#print([titanic.info()])

titanic.describe(include='all')
```

```python
# get number of null values per columns

missing= titanic.isnull().sum()
missing
```

```python
# we can see all missing value in heatmap

plt.figure(figsize=(20,15))
sns.heatmap(titanic.isnull())
```

```python
# we can see numeric column's corr in heatmap

sns.heatmap(titanic.corr(), annot=True)
```

```python
# filter data of columns "sex","pclass","age","fare","survived"
some_columns= titanic[["sex","pclass","age","fare","survived"]]
some_columns
```

```python
# draw graph below

a= sns.PairGrid(some_columns)
a.map_diag(sns.histplot)
a.map_offdiag(sns.scatterplot)

```

```python
sns.pairplot(some_columns)
```

```python
# Change default palette to GREEN and RED

a= sns.PairGrid(some_columns, hue='sex', palette=['green', 'red'])
a.map_diag(sns.kdeplot, shade=True)
a.map_offdiag(sns.scatterplot)
a.add_legend()
```

```python
sns.pairplot(some_columns, hue='sex', palette=['green','red'])
```
